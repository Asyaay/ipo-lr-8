Задание 1: 

```python
data = {
    '100':[6, 9, 7, 10, 7, 9, 5, 3, 6, 6, 9, 10, 5, 6, 5, 6, 8, 6, 7, 5, 5], #7
    '101':[8, 7, 10, 7, 8, 10, 8, 9, 8, 8, 9, 9, 9, 8, 8], #5
    '102':[7, 8, 9, 4, 10, 8, 10, 7, 9, 9, 9, 9, 5, 9 ,8, 9, 9, 10, 9, 7, 9, 9, 9, 9, 4, 10, 9], #9
    '103':[6, 4, 8, 3, 10, 7, 9, 9, 10, 9, 4, 5, 10, 4, 5], #5
    '104':[9, 6, 8, 7, 5, 9, 6, 4, 9, 9, 3, 7, 5, 8, 8, 9, 8, 7, 4, 7, 9, 7, 8, 9, 4, 8, 9, 6, 7, 9], #10
} #словарь
score = {group: sum(scores) / len(scores) for group, scores in data.items()} 
#с помощью генератора словарей проходимся по группам и отметкам, складываем отметки,
#находим кол-во отметок и с помощью деления находим среднее арифметическое отметок и сохраняем в словаре   
groups = list(score.keys()) #создаем список, в котором будут номера групп (ключи)
scores = list(score.values()) #создаем список, в котором будут баллы (значения)
print(f"Средние баллы по группам: {score}") #выводим результаты
```

Задание 2:

```python
import matplotlib.pyplot as plt #подключаем библиотеку
plt.figure(figsize=(10, 6)) #создаем пустой лист (задаем ширину и высоту)
plt.bar(groups, scores, color='mediumpurple') #создаем диаграмму с помощью
#значений, которые мы получили ранее, а так же задаем цвет столбов 
```

```python
import matplotlib.pyplot as plt #подключаем библиотеку
plt.figure(figsize=(10, 6)) #создаем пустой лист (задаем ширину и высоту)
plt.bar(groups, scores, color='mediumpurple') #создаем диаграмму с помощью
#значений, которые мы получили ранее, а так же задаем цвет столбов 
plt.xlabel('Группы') #задаем название оси х
plt.ylabel('Средний балл') #задаем название оси у
plt.title('Средние баллы по группам') #задаем название диаграммы
plt.show() #создаем график
```

Задание 4:

```python
from openpyxl import Workbook #подключаем библиотеку
wb = Workbook() #создаем Excel файл
ws = wb.active #выбираем лист, на котором будет выводиться информация
ws.title = "NewPage" #называем лист
ws.sheet_properties.tabColor = "1072BA" #выбираем цвет листа
file_name = 'students_result.xlsx' #называем сам файл
groups = list(score.keys()) #создаем список, в котором будут номера групп (ключи)
scores = list(score.values()) #создаем список, в котором будут баллы (значения)
ws['A1'] = 'Группа' #в А ячейке 1 стоблца пишем навание столбца
ws['B1'] = 'Средний балл' #в В ячейке й столбца пишем название столбца
for i in range(len(groups)): #перебор значений из количества групп
    ws.cell(row = i + 2, column = 1, value = groups[i]) #начиная со второй строки
#записываем значения из группы в первый столбик
    ws.cell(row = i + 2, column = 2, value = scores[i]) #начиная со второй строки
#записываем значения из баллов во второй столбик
    wb.save(file_name) #сохраняем файл
```
